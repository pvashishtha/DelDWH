```python
import requests
```

```python
# GET/POST
```

```python
response=requests.get('https://catfact.ninja/fact')
```

```python
response
```

```python
# 200- OK, 400+ - Resource at URL is not found
# 500+ - Server is not available
response.status_code
```

```python
response.text
```

```python
response.json()
```

```python
response.json()['fact']
```

```python
info=response.headers
info['server']
```

```python
if response.status_code==200:
    print(response.json()['fact'])
else:
    print('No Response')
```

```python
# https://api.nationalize.io?name=prateek
response=requests.get('https://api.nationalize.io?name=Dawn')
if response.status_code==200:
    d=response.json()
    print(f"Country Code: {d['country'][0]['country_id']}")
    print(f"Probability: {d['country'][0]['probability']}")
else:
    print('No Response')
```

```python
# https://api.coindesk.com/v1/bpi/currentprice.json
import time
for i in range(6):
    response=requests.get('https://api.coindesk.com/v1/bpi/currentprice.json')
    d=response.json()
    coin=d['chartName']
    price=d['bpi']['USD']['rate']
    print(f'{coin} rate today is $ {price}')
    time.sleep(10)
```
