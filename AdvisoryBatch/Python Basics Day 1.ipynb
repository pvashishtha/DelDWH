### Flow Control

#### Conditional Statements

```python
# if <boolean-expression>:

if a<b:
    # line1 # begin of block
    # line2
    # line 3  # cont to block
# line 4  # block ended- this is outside if
```

```python
a=10
b=50
if a>b:
    print('if starts')
    print('a is bigger number')
    print('if-ends here')
print('this is outside if')
```

```python
# if-else
a=10
b=50
if a<b:
    print('a is smaller')
else:
    print(' b is smaller')
print('Outside if-else, always printed!')
```

```python
# multiple if statements
s='abc123'
if s.isalpha():
    print('Alphabets only')
if s.isdecimal():
    print('Decimal numbers only')
if s.isalnum():
    print('alphanumeric string')
```

```python
# if-elif-else ladder
x,y=10,20
if x+y>0 and x+y<100:
    print('Small Numbers')
elif x+y>100 and x+y<1000:
    print('Moderate Numbers')
elif x+y>1000 and x+y<10000:
    print('Average Value Numbers')
else:
    print('High Value Numbers')
```

```python
# write a code to find largest of 3 numbers
a,b,c=1000,2000,15000

if a>b and a>c:
    print('A largest')
elif b>a and b>c:
    print('B is largest')
else:
    print('C is largest')
```

```python
# Write a code to check whether number entered by user is MAGIC or
# NON-MAGIC number.
# A magic number is one that can be divisible by both 4 and 17.

num=int(input())
if num%4==0 and num%17==0:
    print('MAGIC')
else:
    print('NON-MAGIC')
```
