```python
x=90
if x>90:
    print('Class 1')
elif x>80 and x<=90:
    print('Class 2')
elif x>60 and x<=80:
    print('Class 3')
else:
    print('Class 4')
```

```python
i=0
while i<5:
    print('i=', i)
    i+=1
```

```python
while True:
    print('i=', i, end=' ')
    if i%17==0:
        break
    else:
        print('next iteration...')
        i+=1
```

```python
i=0
while i<20:
    i+=1
    if i%2==0:
        continue
    else:
        print('i=', i, end=' ')
        print('next iteration...')
```
