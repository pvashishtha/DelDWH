```python
import random



def guessing():
    file=open('words.txt','r') #file containing list of words
    data=file.read()
    global li
    li=data.split('\n')
    #print(li)



def chooseWord():
    global theWord
    theWord=random.choice(li)

   
def inputGuess():
    a=input('Enter an alphabet: ')
    b=a.lower()
    if a.isalpha():
        return b  
    else:
        print("That wasn't an alphabet!!!! Enter a valid character")
        inputGuess()
        



print('Welcome to GUESS THE WORD!!')
name=input('Enter your Name: ')
Instructions='''Guess the 5 character word in 10 Tries!!
'''
print('Hi !! '+name)
print(Instructions)



guessing()
chooseWord()



word=list()
for x in theWord:
    word.append(x)
#print(word)
g=['-','-','-','-','-']

for i in range(10):
    ltr=str(inputGuess())
    print(ltr)
    if ltr in theWord:
        for x in range(0,len(word)):
            #print(word[x])
            if(word[x]==ltr):
                g[x]=ltr
        print('Right Guess!!')
        print('Tries remaining: ',10-i-1)
    else:
        print('Wrong Guess')
        print('Tries remaining: ',10-i-1)       
    print(''.join(g))
    if(g==word):
        print('''
             
        
        ''')
        print('Congratulationss!!You have won')
        break
print('GAME OVER!!')
print('The correct word is:',theWord)
```
