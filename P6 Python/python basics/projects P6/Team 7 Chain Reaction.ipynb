```python
from IPython.display import display,clear_output
from colorama import Fore, Back, Style
```

```python
def chain_reaction():
    w=0
    print('Welcome to Chain Reaction!')
    print('Sorry for the inconvience, The game is under development, So only 2 players can play')
    print('The first player is red\nThe seconde player is green')
    a=[[' ',0],[' ',0],[' ',0],[' ',0],[' ',0],[' ',0],[' ',0],[' ',0],[' ',0]]
    board_print(a) # displays the grid
    red_input(a)   # red's input
    clear_output() # clearing the output screen
    board_print(a)
    green_input(a) # green's input
    clear_output()
    board_print(a)
    w=winner_check(a) # check for a winner
    while(w!=1): # loop goes on untill we find a winner
        red_input(a)
        clear_output()
        board_print(a)
        w=winner_check(a)
        if w == 1:    # breaks the loop when we find a winner
            break
        green_input(a)
        clear_output()
        board_print(a)
        w=winner_check(a)
    print('\nGame ended successfully!')    
```

```python
chain_reaction()
```

```python
def red_input(a):
    r = int(input(Fore.RED+Style.BRIGHT+"Its red's turn,Enter the position you want put the ball" ))
    if r not in range(1,10):
        print(Fore.RESET+Style.BRIGHT+'Invalid position, Enter 1-9')
        red_input(a)
    else:
        if a[r-1][0]=='G':
            print(Fore.RESET+Style.BRIGHT+'\nYou cant enter in a green box!')
            red_input(a)
        else:
            a[r-1][1] = a[r-1][1] + 1
            a[r-1][0] = 'R'
            checks(a)
```

```python
def green_input(a):
    g = int(input(Fore.GREEN+Style.BRIGHT+ "Its green's turn,Enter the position you want put the ball" ))
    if g not in range(1,10):
        print(Fore.RESET+Style.BRIGHT+'Invalid position, Enter 1-9')
        green_input(a)
    else:
        if a[g-1][0]== 'R':
            print(Fore.RESET+Style.BRIGHT+'\nYou cant enter in a red box!')
            green_input(a)
        else:
            a[g-1][1] = a[g-1][1] + 1
            a[g-1][0] = 'G'
            checks(a)
```

```python
def corner_check(a):
    ec = 0
    if a[0][1] == 2:
        a[0][1] = 0
        a[1][1] += 1
        a[1][0] = a[0][0]
        a[3][1] += 1
        a[3][0] = a[0][0]
        a[0][0] = ' '
        ec = 1
    elif a[2][1] == 2:
        a[2][1] = 0
        a[1][1] += 1
        a[1][0] = a[2][0]
        a[5][1] += 1
        a[5][0] = a[2][0]
        a[2][0] = ' '
        ec = 1
    elif a[6][1] == 2:
        a[6][1] = 0
        a[3][1] += 1
        a[3][0] = a[6][0] 
        a[7][1] += 1
        a[7][0] = a[6][0]
        a[6][0] = ' '
        ec = 1
    elif a[8][1] == 2:
        a[8][1] = 0
        a[3][1] += 1
        a[3][0] = a[8][0]
        a[7][1] += 1
        a[7][0] = a[8][0]
        a[8][0] = ' '
        ec = 1
    if ec == 1:
        return True
    if ec == 0:
        return False
```

```python
def middle_check(a):
    em = 0
    if a[4][1] == 4:
        a[4][1] = 0
        a[1][1] += 1
        a[1][0] = a[4][0]
        a[3][1] += 1
        a[3][0] = a[4][0]
        a[5][1] += 1
        a[5][0] = a[4][0]
        a[7][1] += 1
        a[7][0] = a[4][0]
        a[4][0] = ' '
        em = 1
    if em == 1:
        return True
    if em == 0:
        return False     
```

```python
def edge_check(a):
    flg=False
    if a[1][1]==3: #edge 0
        a[0][1]+=1
        a[0][0]=a[1][0] #updating user of adjacent cell to user of corner 0 cell if burst
        
        a[2][1]+=1
        a[2][0]=a[1][0]
        
        a[4][1]+=1
        a[4][0]=a[1][0]
        
        a[1][1]=0
        a[1][0] = ' '
        flg=True
        
    if a[3][1]==3: #edge 2
        a[0][1]+=1
        a[0][0]=a[3][0] #updating user of adjacent cell to user of corner 2 cell if burst
        
        a[4][1]+=1
        a[4][0]=a[3][0]
        
        a[6][1]+=1
        a[6][0]=a[3][0]
        
        a[3][1]=0
        a[3][0] = ' '
        flg=True
        
    if a[5][1]==3: #edge 6
        a[2][1]+=1
        a[2][0]=a[5][0]
        
        a[4][1]+=1
        a[4][0]=a[5][0]
        
        a[8][1]+=1
        a[8][0]=a[5][0]
        
        a[5][1]=0
        a[5][0] = ' '
        flg=True





    if a[7][1]==3: #edge 8
        a[4][1]+=1
        a[4][0]=a[7][0]
        
        a[6][1]+=1
        a[6][0]=a[7][0]
        
        a[8][1]+=1
        a[8][0]=a[7][0]
        
        a[7][1]=0
        a[7][0] = ' '
        flg=True
        
    return flg
```

```python
def checks(a):
    while(corner_check(a) != False or edge_check(a) != False and middle_check(a) != False):
        corner_check(a) 
        edge_check(a)
        middle_check(a)
```

```python
def winner_check(a):
    if 'R' not in [a[0][0],a[1][0],a[2][0],a[3][0],a[4][0],a[5][0],a[6][0],a[7][0],a[8][0]]:
        print(Fore.GREEN+Style.BRIGHT+'\nGreen is the winner!')
        w=1
    elif 'G' not in [a[0][0],a[1][0],a[2][0],a[3][0],a[4][0],a[5][0],a[6][0],a[7][0],a[8][0]]:
        print(Fore.RED+Style.BRIGHT+'\nRed is the winner!')
        w=1
    else:
        w=0
    return w
```
