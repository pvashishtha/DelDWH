```python
# Working with XML Files
```

```python
import xml.etree.ElementTree as ET
```

```python
tree=ET.parse("guest.xml")
```

```python
root=tree.getroot()
root
```

```python
from colorama import Fore, Style
f='First Name'
l='Last Name'
p='Phone'
print(Fore.RED+f'{f:<16}{l:<16}{p:<16}')
print(Style.RESET_ALL)
for item in root.iter('guestbook'):
    for name in item.iter('guest'):
        for fname, lname, phone in zip(name.iter('fname'),
                                      name.iter('lname'),
                                      name.iter('phone')):
            print(f'{fname.text:<16}{lname.text:<16}{phone.text:<16}')
```

```python
from bs4 import BeautifulSoup as bs
```

```python
with open('guest.xml', 'r') as file:
    data=file.read()
all_data=bs(data, 'xml')
tag_data=all_data.find('guest')
print(tag_data)
```

```python
fname=all_data.find_all('fname')
lname=all_data.find_all('lname')
phone=all_data.find_all('phone')
```

```python
print(Fore.RED+f'{f:<16}{l:<16}{p:<16}')
print(Style.RESET_ALL)
for fname, lname, phone in zip(fname, lname, phone):
    print(f'{fname.text:<16}{lname.text:<16}{phone.text:<16}')
```

```python
# List Buckets
response=client.list_buckets()

print('Listing Buckets...')
for bucket in response['Buckets']:
    print(f'-- {bucket["Name"]}')
```

```python
# Uploading file
def upload_files(file_name, bucket, object_name=None, args=None):
    if object_name==None:
        object_name=file_name
    client.upload_file(file_name, bucket, object_name, ExtraArgs=args)
    print(f'{file_name} has been uploaded to {bucket}')
    
    
upload_files('data.csv', 'p4abucket')
upload_files('data.json', 'p4abucket')
upload_files('guest.xml', 'p4abucket')
```

```python
# Download File
s3_resource=boto3.resource("s3",
                          aws_secret_access_key=SECRET_KEY,
                           aws_access_key_id=ACCESS_KEY,
                           region_name=AWS_REGION)

s3_object=s3_resource.Object('p4abucket','panda.jpeg')
s3_object.download_file('panda_new.jpg')
print('File Downloaded!')
```
