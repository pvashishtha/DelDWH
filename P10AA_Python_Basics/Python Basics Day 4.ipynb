##### Question:
Write a function rotate(list) that rotates only negative numbers, when encountered, from any location in list to end of list, maintaining the order of appearance in original list. After all rotations, all negative numbers must follow all positive numbers.

Example: Input=> li=[5,2,3,7,-1,4,-3,-5,2,1,8,9,-7]
         Output=> [5,2,3,7,4,2,1,8,9,-1,-3,-5,-7]

```python
def rotate(li):
    pos,neg=[],[]
    for x in li:
        pos.append(x) if x>=0 else neg.append(x)
    return pos+neg
rotate([5,2,3,7,-1,4,-3,-5,2,1,8,9,-7])
```

```python
def rotate(li):
    i=0
    l=len(li)
    while i<l:
        if li[i]<0:
            li.append(li[i])
            li.pop(i)
            l-=1
            if i>0:
                i-=1
        i+=1
    return li
rotate([5,2,3,7,-1,4,-3,-5,2,1,8,9,-7])
```

#### Question:
Write a function code as alternating(list) that finds if the elements in the list increases or dcreases in alternating manner. If its has alternating pattern (either increase or decrease) then return "True" otherwise return "False".

Example: 

Input=> [3,1,4,2,7,5,9] Output=> TRUE
         
Input=> [4,3,7,2,2,8,1] Output=> FALSE

```python
def alternating(li):
    for i in range(len(li)-2):
        s1=li[i]==li[i+1]
        s11=li[i+1]==li[i+2]
        s2=li[i]<li[i+1]<li[i+2]
        s3=li[i]>li[i+1]>li[i+2]
        if any([s1,s11,s2,s3]):
            return False
    return True
print(alternating([3,1,4,2,7,5,5]))
print(alternating([4,3,7,2,2,8,1]))
```

```python
# copy binary large file by parts
def file_copy(file1, file2):
    fp1=open(file1, 'rb')
    fp2=open(file2,'wb')
    fp1.seek(0,2)
    file_size=fp1.tell()
    print('original File Size: ',file_size)
    fixed_partition=file_size//50000
    remaining_partition_size=file_size-(fixed_partition*50000)
    print('Fixed Partitions: ', fixed_partition)
    print('Last Chunk Size: ',remaining_partition_size)
    fp1.seek(0)
    for it in range(fixed_partition+1):
        if it==fixed_partition:
            if remaining_partition_size==0:
                break
            else:
                data=fp1.read(remaining_partition_size)
                fp2.write(data)
                fp2.flush()
                print('*** writing last chunk ***')
        else:
            data=fp1.read(50000)
            fp2.write(data)
            fp2.flush()
            print('*** writing full chunk ***')
    fp1.close()
    fp2.close()
    print('File Copy Ended!')

file_copy('panda.jpg', 'newPanda.jpg')
```

### Exception Handling

```python
# try - except

x=int(input())
y=int(input())
try:
    div=x/y
    print(div)
except:
    print("Y can't be zero")
print('End')
```

```python
# Multiple Except Blocks with named Exceptions
try:
    x=int(input())
    y=int(input())
    file=open('myfile5.txt', 'r')
    file.close()
    print(x/y)
except ZeroDivisionError:
    print('Error Occured: Zero Division')
except FileNotFoundError:
    print('Error Occured: File Not Found')
except:
    print('Unknown Error')
print('END!')
```

```python
# else and finally
try:
    x=int(input())
    y=int(input())
    file=open('myfile.txt', 'r')
    print(x/y)
except ZeroDivisionError:
    print('Error Occured: Zero Division')
except FileNotFoundError:
    print('Error Occured: File Not Found')
except:
    print('Unknown Error')
else:
    print('Try run successful')
finally:
    file.close()
    print('file closed')
print('END!')
```

```python
# exception arguments
try:
    x=int(input())
    y=int(input())
    print(x/y)
except ZeroDivisionError as e:
    print(e)
```

```python
# consolidated exceptions
try:
    x=int(input())
    y=int(input())
    file=open('myfile.txt', 'r')
    print(x/y)
except (ZeroDivisionError, FileNotFoundError):
    print('Known Error')
except:
    print('Unknown Error')
```

```python
# User Defined Exceptions
class MyExceptionError(Exception):
    def __init__(self,msg):
        self.msg=msg
        return None
```

```python
try:
    file=open('myfile.txt','w')
    text=input('Enter Your Name: ')
    if text=='':
        raise MyExceptionError("Name can not be Null...")
    file.write(text)
    file.close()
except MyExceptionError as e:
    print('Error: '+e.msg)
    #raise
except:
    print('Unknown Error')
```

```python
# Write a function that generates a N character captcha that contains
# combination of Uppercase letter, lowercase letter, numbers
def captcha(N):
    import random
    cap=''
    A=[chr(x) for x in range(65,91)]
    a=[chr(x) for x in range(97,123)]
    n=[str(x) for x in range(10)]
    collection=[A,a,n]
    for _ in range(N):
        idx=random.choice([0,1,2])
        cap+=random.choice(collection[idx])
    return cap
```

```python
captcha(6)
```

```python
# for a given list, find all the pairs of elements whose difference
# is an odd number.

li=[3,6,1,7,9,5,2,6]
s=set()
for i in li:
    for j in li:
        if abs(i-j)%2!=0:
            if i<j:
                s.add((i,j))
            else:
                s.add((j,i))
print(s)
```
